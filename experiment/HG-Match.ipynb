```python
import pandas as pd 
import numpy as np
import os
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.rcParams['savefig.dpi'] = 300 
plt.rcParams['figure.dpi'] = 300
```

```python
# the folder that contain the figures
FIG_DIR = 'hg'

path = os.path.join(os.getcwd(), FIG_DIR)
# check whether the directory exist or not
if not os.path.exists(path):
    os.mkdir(path)

def plot_line(a, data,COLOR,LINESTYLE,MAKER,LABEL):
    plt.plot(a, data,color=COLOR,linestyle=LINESTYLE, 
             markersize=15, 
             markeredgewidth =1.5,
             markerfacecolor='none',
             marker=MAKER,label = LABEL)
```

```python
import numpy as np

filename = os.path.join(path, 'cp.pdf')


fig = plt.figure(1, figsize=(6.8, 4.0))
ax = fig.add_subplot(111)


plt.style.use('default')
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams["legend.framealpha"] = 0
plt.rcParams["figure.figsize"] = (10,3)
plt.rcParams["legend.handletextpad"] = 0.1
plt.rcParams["legend.columnspacing"] = 0.3


default=[
    ['$q_2$',0.00091254945, 0.001965, 0.001755, 0.006165],
    ['$q_3$',2.26E-04, 0.001995, 0.004235, 0.008525],
    ['$q_4$',0.445663968, 1.41718, 3.75891, 12.08624],
    ['$q_6$',2.493993383, 9.746605, 30.62805, 228.41897]
]

default_df=pd.DataFrame(default)

index_x = default_df[0].tolist()

HG = default_df[1].tolist()
CECI = default_df[2].tolist()
CFL = default_df[3].tolist()
DP = default_df[4].tolist()

x = np.arange(len(index_x))
# x =list(range(len(our_list)))
total_width, n = 1.1, 4
width = total_width / (n+1)

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x, DP, width=width, linewidth=0.5,color="w" ,edgecolor="black")

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x+width, CFL, width=width,linewidth=0.5,edgecolor='black',hatch="/"*2,
        fc = 'w')

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x+width+width, CECI, width=width,linewidth=0.5,edgecolor='black',hatch="\\\\",
        fc = 'w')

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x+width+width+width, HG, width=width,linewidth=0.5,edgecolor='black',
        fc = 'black')


plt.xticks(x+1*width, index_x)


plt.yscale('log')
plt.ylim(0.0001,1*pow(10,3))
plt.ylabel("Time Cost (sec)",fontsize = 20)


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


# plt.legend(fontsize=16 ,ncol=4,loc="upper left", handlelength = 1.8, handletextpad = 0.4, labelspacing = 0, columnspacing=1) 
plt.savefig(filename, bbox_inches='tight')
plt.show()
```
