```python
import pandas as pd 
import numpy as np
import os
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.rcParams['savefig.dpi'] = 300 
plt.rcParams['figure.dpi'] = 300
```

```python
# the folder that contain the figures
FIG_DIR = 'patmat'

path = os.path.join(os.getcwd(), FIG_DIR)
# check whether the directory exist or not
if not os.path.exists(path):
    os.mkdir(path)

def plot_line(a, data,COLOR,LINESTYLE,MAKER,LABEL):
    plt.plot(a, data,color=COLOR,linestyle=LINESTYLE, 
             markersize=15, 
             markeredgewidth =1.5,
             markerfacecolor='none',
             marker=MAKER,label = LABEL)
```

```python
data="""GO	38	40	120
LJ	278	364	455
OR	759	918	1044
UK	1888	2139	2519
FS	12901	14280	15462"""

index = 0
for _, line in enumerate(data.split('\n')):
    n = line.split('	')[0]
    line = line.split('	')[1:]
#     print(line)
#     if float(line[2]) > float(line[4]):
    line = ",".join(line)
    print ("[\"{}\",{}],".format(n, line))
    index += 1
    
```

```python
import numpy as np

filename = os.path.join(path, 'dsize.pdf')


fig = plt.figure(1, figsize=(8.8, 4.0))
ax = fig.add_subplot(111)


plt.style.use('default')
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams["legend.framealpha"] = 0
plt.rcParams["figure.figsize"] = (10,3)
plt.rcParams["legend.handletextpad"] = 0.1
plt.rcParams["legend.columnspacing"] = 0.3


default=[
    ["GO",38,40,120],
    ["LJ",278,364,455],
    ["OR",759,918,1044],
    ["UK",1888,2139,2519],
    ["FS",12901,14280,15462],
]

default_df=pd.DataFrame(default)

index_x = default_df[0].tolist()

PatmatX = default_df[1].tolist()
PatMatXn = default_df[2].tolist()
DualSim = default_df[3].tolist()

x = np.arange(len(index_x))
total_width, n = 1.0, 3
width = total_width / (n+1)

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x, PatmatX, width=width, linewidth=0.5,color="w" ,edgecolor="black", label="PatmatX")

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x+width, PatMatXn, width=width,linewidth=0.5,edgecolor='black',hatch="/"*2,
        fc = 'w', label="PatMatX-nopack")

for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x+width + width, DualSim, width=width,linewidth=0.5,edgecolor='black',hatch="\\"*2,
        fc = 'w', label="DualSim")


# set the y | position
plt.xticks(x+1*width, index_x)


plt.yscale('log')
# plt.minorticks_off()

# index_y = [1*pow(10,0), 1*pow(10,1), 1*pow(10,2), 1*pow(10,3), 1*pow(10,4), 1*pow(10,5)]
# y_labels = [r'$10^0$', r'$10^1$', r'$10^2$', r'$10^3$', r'$10^4$', 'OT']
# plt.yticks(index_y, y_labels)
plt.ylim(10,1*pow(10,5))

plt.ylabel("On-disk Size (MB)",fontsize = 20)


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
# plt.xlabel(r"$q_1$",fontsize=20)


plt.legend(fontsize=18 ,ncol=4,loc="upper left", handlelength = 1.8, handletextpad = 0.4, labelspacing = 0, columnspacing=1) 
plt.savefig(filename, bbox_inches='tight')
plt.show()
```
